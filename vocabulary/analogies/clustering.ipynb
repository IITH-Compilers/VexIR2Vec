```python
fname = "/path/to/seedEmbedding.txt"
f = open(fname, "r")
seedEmbedding = {}
X = []
Y = []

for line in f.readlines():
    embedding = []
    for i in line.split(':')[1].split(','):
        if i[0] == '[':
            embedding.append(float(i[1:]))
        elif i[-1] == ']':
            embedding.append(float(i[:-1]))
        elif i == '\n':
            continue
        else:
            embedding.append(float(i))
    embedding = np.array(embedding)
    seedEmbedding[line.split(':')[0]] = embedding
    X.append(embedding)
    Y.append(line.split(':')[0])
```

```python
print(seedEmbedding)
```

```python
X = np.array(X)
print(X)
```

```python
rev_Y = {}
for i in range(len(Y)):
    rev_Y[Y[i]] = i
print(rev_Y)
```

```python
fp = open('/path/to/vexOps.txt', 'w')
for key in rev_Y:
    # print(key)
    fp.write(key+'\n')
fp.close()
```

```python

X_embedded = TSNE(n_components=2, learning_rate='auto', perplexity=7, n_iter=10000, early_exaggeration=8, init='pca').fit_transform(X)
```
